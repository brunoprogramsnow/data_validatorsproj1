```python
import requests
import json
import pandas as pd
import numpy as np
import import_ipynb
from matplotlib import pyplot as plt
from pprint import pprint
from Zillow_CSV_Exploration_and_Cleanup import Top20cities_df
import time
from keys import gkey
```

```python
# Import city and state from other ipynb notebook
lat_long_df = Top20cities_df

# Create new columns to add lat and lng
lat_long_df["Lat"] = ""
lat_long_df["Lng"] = ""
lat_long_df["bus_station"] = ""
lat_long_df["train_station"] = ""
lat_long_df["transit_station"] = ""
lat_long_df["subway_station"] = ""
lat_long_df["airport"] = ""
lat_long_df["taxi_stand"] = ""
lat_long_df
```

```python
# Create a params dict
params ={
    "key": gkey
}

# Loop through lat_long_df to grab values for the lat_lng search
for index, row in lat_long_df.iterrows():
    
    # Base_url for google maps api
    base_url = "https://maps.googleapis.com/maps/api/geocode/json"
    
    # Creates two vars that reads the values in the respective rows
    city = row['CITY']
    state = row['STATE']
    
    # Adds address to the params dict
    params["address"] = f"{city},{state}"
    
    lat_lng_request = requests.get(base_url, params= params)
    
    lat_lng_request = lat_lng_request.json()

    
    lat_long_df.loc[index, "Lat"] = lat_lng_request["results"][0]["geometry"]["location"]["lat"]
    lat_long_df.loc[index, "Lng"] = lat_lng_request["results"][0]["geometry"]["location"]["lng"]
    
lat_long_df
```

```python
# Set initial variables
page_token = ""
target_radius = 10000
```

```python
# Airport data

# Params dictionary to update each iteration
params = {
    "radius": target_radius,
    "types": "airport",
    "key": gkey,
    "pagetoken" : page_token
}

for index, row in lat_long_df.iterrows():
    
    lat = row["Lat"]
    lng = row["Lng"]
       
    city = row['CITY']
    state = row['STATE']
    
    params["location"] = f"{lat},{lng}"

    base_url = "https://maps.googleapis.com/maps/api/place/nearbysearch/json"

    new_request = requests.get(base_url, params=params)
    
    new_request = new_request.json()

    # Get first search page results
    search_result_num = len((new_request["results"]))

    # Google Places API generates up to 20 results per page.
    # Google Places API on free account only allows maximum of 60 results.

    # Loops if there is a next_page_token available, ends the search if there is no next_page_token
    while True:
        try:
            
            params["pagetoken"] = new_request['next_page_token']
            time.sleep(2)
            new_request = requests.get(base_url, params=params).json()
            search_result_num += len((new_request["results"]))

        except KeyError:
            
            # Reset pagetoken
            params["pagetoken"] = ""
            
            print(f"No next page token. Search is complete for {city}, {state}.")
            break    
    
    lat_long_df.loc[index, "airport"] = search_result_num
    
    print(f"{search_result_num} airports")

print("All searches completed.")
```
